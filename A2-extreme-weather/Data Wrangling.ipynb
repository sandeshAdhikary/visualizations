```python
import pandas as pd
import numpy as np
import json
from urllib.request import urlopen
import time
```

```python
start_year = 1980
end_year = 2021
disasters = ['drought','flooding','freeze',
             'severe-storm','tropical-cyclone',
             'wildfire','winter-storm','all-disasters']
years = list(range(start_year,end_year,1))
```

# USA Data

```python
data = pd.DataFrame([])
# Load json data
with open('data/usa_data.json', 'r') as j:
    json_data = json.loads(j.read())
data = pd.DataFrame([])
for disaster in disasters:
    disaster_data = {x:json_data['data'][str(x)][disaster] for x in years}
    disaster_data = pd.DataFrame.from_records(disaster_data).transpose()
    disaster_data.reset_index(level=0,inplace=True)
    disaster_data.rename(columns={'index':'year'}, inplace=True)
    disaster_data['disaster'] = disaster
    data = data.append(disaster_data,ignore_index=True, sort=False)
    
    
    

#     region_data['region'] = region
```

```python
data.to_csv('data/usa_data.csv', index=False)
```

### Combine JSON files into single dataframe

```python
data = pd.DataFrame([])
for region in regions:
    print('Processing region: {}'.format(region))
    # Load json data
    with open('data/{}_data.json'.format(region), 'r') as j:
        region_json = json.loads(j.read())
    region_data = pd.DataFrame([])
    for disaster in disasters:
        disaster_data = {x:region_json['data'][str(x)][disaster] for x in years}
        disaster_data = pd.DataFrame.from_records(disaster_data).transpose()
        disaster_data.reset_index(level=0,inplace=True)
        disaster_data.rename(columns={'index':'year'}, inplace=True)
        disaster_data['disaster'] = disaster
        region_data = region_data.append(disaster_data,ignore_index=True, sort=False)
    

    region_data['region'] = region
#     # Append dataframe to data
    data = data.append(region_data, ignore_index=True)
```
